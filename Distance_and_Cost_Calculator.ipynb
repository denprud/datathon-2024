```python
# Read the CSV file
df3 = pd.read_csv('all.csv')
df4 = pd.merge(df3, df2, on='geography')

def get_coordinates(geoID):
    # Convert geoID to int or float if necessary
    try:
        geoID = int(geoID)
    except ValueError:
        try:
            geoID = float(geoID)
        except ValueError:
            pass

    row = df4[df4['geography'] == geoID]


    if len(row) > 0:
        return row['tl_2023_48_tract.INTPTLAT'].values[0], row['tl_2023_48_tract.INTPTLON'].values[0]
    else:
        return None

# User input
geoID = input("Please enter a geoID: ")

# Get coordinates
coordinates = get_coordinates(geoID)

if coordinates:
    print(f"The coordinates for geoID {geoID} are {coordinates}.")
else:
    print(f"No coordinates found for geoID {geoID}.")
```

```python
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    r = 6371  # Radius of earth in kilometers
    return c * r
```

```python
# User's coordinates
user_lat, user_lon = coordinates  # replace with actual coordinates

# Initialize variables to store the minimum distance and index
min_distance = np.inf
min_index = None

# Loop through the DataFrame
for i, row in df.iterrows():
    # Calculate the distance
    distance = haversine(user_lat, user_lon, row['tl_2023_48_tract.INTPTLAT'], row['tl_2023_48_tract.INTPTLON'])

    # If the distance is less than the current minimum, update the minimum distance and index
    if distance < min_distance:
        min_distance = distance
        min_index = i

# Print the index of the closest coordinates
print(f"The closest coordinates are at index {min_index} with a distance of {min_distance} kilometers.")
```
