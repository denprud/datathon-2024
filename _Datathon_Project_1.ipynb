```python
TX_FEMA['geography'] = TX_FEMA['geography'].str[-11:]
TX_pov['geography'] = TX_pov['geography'].str[-11:]
TX_fin['geography'] = TX_fin['geography'].str[-11:]
TX_income['geography'] = TX_income['geography'].str[-11:]
TX_housing['geography'] = TX_housing['geography'].str[-11:]

GA_FEMA['geography'] = GA_FEMA['geography'].str[-11:]
GA_pov['geography'] = GA_pov['geography'].str[-11:]
GA_fin['geography'] = GA_fin['geography'].str[-11:]
GA_income['geography'] = GA_income['geography'].str[-11:]
GA_housing['geography'] = GA_housing['geography'].str[-11:]

FL_FEMA['geography'] = FL_FEMA['geography'].str[-11:]
FL_pov['geography'] = FL_pov['geography'].str[-11:]
FL_fin['geography'] = FL_fin['geography'].str[-11:]
FL_income['geography'] = FL_income['geography'].str[-11:]
FL_housing['geography'] = FL_housing['geography'].str[-11:]

TX_extra['geography'] = TX_extra['geography'].str[-11:]
GA_extra['geography'] = GA_extra['geography'].str[-11:]
FL_extra['geography'] = FL_extra['geography'].str[-11:]
```

```python
TX = pd.merge(TX_FEMA, TX_pov, on='geography', suffixes=('_fema', '_pov'))
TX = pd.merge(TX, TX_fin, on='geography', suffixes=('_pov', '_fin'))
TX = pd.merge(TX, TX_housing, on='geography', suffixes=('_fin', '_housing'))
TX = pd.merge(TX, TX_income, on='geography', suffixes=('_housing', '_income'))
TX = pd.merge(TX, TX_extra, on = 'geography', suffixes=('_income','_extra'))

GA = pd.merge(GA_FEMA, GA_pov, on='geography', suffixes=('_fema', '_pov'))
GA = pd.merge(GA, GA_fin, on='geography', suffixes=('_pov', '_fin'))
GA = pd.merge(GA, GA_housing, on='geography', suffixes=('_fin', '_housing'))
GA = pd.merge(GA, GA_income, on='geography', suffixes=('_housing', '_income'))
GA = pd.merge(GA, GA_extra, on = 'geography', suffixes=('_income','_extra'))

FL = pd.merge(FL_FEMA, FL_pov, on='geography', suffixes=('_fema', '_pov'))
FL = pd.merge(FL, FL_fin, on='geography', suffixes=('_pov', '_fin'))
FL = pd.merge(FL, FL_housing, on='geography', suffixes=('_fin', '_housing'))
FL = pd.merge(FL, FL_income, on='geography', suffixes=('_housing', '_income'))
FL = pd.merge(FL, FL_extra, on = 'geography', suffixes=('_income','_extra'))

TX_units = TX_housing[['geography', 'occupied_housing_units']]
GA_units = GA_housing[['geography', 'occupied_housing_units']]
FL_units = FL_housing[['geography', 'occupied_housing_units']]

TX = pd.merge(TX, TX_units, on='geography')
GA = pd.merge(GA, GA_units, on='geography')
FL = pd.merge(FL, FL_units, on='geography')

TX = pd.merge(TX,TX_crime, on= 'COUNTY')
GA = pd.merge(GA,GA_crime, on= 'COUNTY')
FL = pd.merge(FL,FL_crime, on= 'COUNTY')


```

```python
TX_df2.replace(to_replace='-', value=0.0, inplace=True)
TX_df2 = TX_df2.replace('[^0-9.]', '', regex=True)

GA_df2.replace(to_replace='-', value=0.0, inplace=True)
GA_df2 = GA_df2.replace('[^0-9.]', '', regex=True)

FL_df2.replace(to_replace='-', value=0.0, inplace=True)
FL_df2 = FL_df2.replace('[^0-9.]', '', regex=True)


TX_df2 = TX_df2.astype(float)
GA_df2 = GA_df2.astype(float)
FL_df2 = FL_df2.astype(float)

# Display the DataFrame to verify the conversion
#print(TX_df2)
```
